Credit health is a very complex topic to understand, yet a vital part of life in America.It would be valuable to understand what factors are needed to get a guaranteed credit approval and what factors to avoid. The end goal will be to understand if an applicant will be approved or denied for a credit application.The features of this dataset have been anonymized to protect the privacy.

1. Data Analysis

```
import pandas as pd
# Import numpy
import numpy as np

# Load dataset
c_approval = pd.read_csv("credit_approval.csv", header=None)

# Inspect data
c_approval.head()
```

```
# Print summary 
c_approval_description = c_approval.describe()
print(c_approval_description)

print("\n")

c_approval_info = c_approval.info()
print(c_approval_info)

print("\n")

# Inspect missing values in the dataset
c_approval.tail(15)
```

The features 2, 7, 10 and 14 contain numeric values (of types float64, float64, int64 and int64 respectively) and all the other features contain non-numeric values.
The dataset has missing values The missing values in the dataset are labeled with '?'

2. Data Imputation

2.1 Data Imputation - Handling Missing Values -
Replaced '?' with NaNs

```
# Inspect missing values in the dataset
print(c_approval.tail(15))

# Replace the '?'s with NaN
c_approval = c_approval.replace('?', np.nan)

# Inspect the missing values again
c_approval.tail(15)
```

Ignoring missing values can affect the performance of a machine learning model drastically, While ignoring the missing values our machine learning model may miss out on information about the dataset that may be useful for its training

2.2 Data Imputation Part 2 - Mean Imputation - NaNs 

```
# Impute the missing values using mean imputation
c_approval.fillna(c_approval.mean(), inplace=True)

# Count the number of NaNs in the dataset 
print(c_approval.isnull().sum())
```

The missing values present in the numeric columns is taken care. There are still some missing values to be imputed for columns 0, 1, 3, 4, 5, 6 and 13. All of these columns contain non-numeric data and this why the mean imputation would not work.

We are going to impute these missing values with the most frequent values as present in the respective columns.

2.3 Data Imputation - Non -Numeric Data Imputation

```
# Iterate over each column
for col in c_approval:
    # Check if the column is of object type
    if c_approval[col].dtype == 'object':
        # Impute with the most frequent value
        c_approval = c_approval.fillna(c_approval[col].value_counts().index[0])

# Count the number of NaNs in the dataset
print(c_approval.isna())
```

3. Data Preprocessing

  3.1 Convert the non-numeric data into numeric.

```
# Import LabelEncoder
from sklearn.preprocessing import LabelEncoder

# Instantiate LabelEncoder
le = LabelEncoder()

# Iterate over all the values of each column and extract their dtypes
for col in c_approval:
    # Compare if the dtype is object
    if c_approval[col].dtype=='object':
    # LabelEncoder to numeric transformation
        c_approval[col]=le.fit_transform(c_approval[col])
```

3.2 Data Preprocessing - Split the data into train and test sets.

```

from sklearn.model_selection import train_test_split

# Drop the features 11 and 13 and convert the DataFrame to a NumPy array
c_approval = c_approval.drop([11, 13], axis=1)
c_approval = c_approval.values

X,y = c_approval[:,0:13] , c_approval[:,-1]

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X,
                                y,
                                test_size=0.3,
                                random_state=42)
```

3.3 - Scale the feature values to a uniform range. 

```

# Import MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX_train = scaler.fit_transform(X_train)
rescaledX_test = scaler.fit_transform(X_test)
```

4.Fitting Logistic Regression to the train set

```
# Import LogisticRegression
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()

logreg.fit(rescaledX_train, y_train)
```

We will now evaluate our model on the test set with respect to classification accuracy. But we will also take a look the model's confusion matrix. The confusion matrix helps us to view our model's performance

```
# Import confusion_matrix
from sklearn.metrics import confusion_matrix

y_pred = logreg.predict(rescaledX_test)

# Get the accuracy score of logreg model
print("Accuracy of logistic regression classifier: ", logreg.score(rescaledX_test, y_test))

print(confusion_matrix(y_test, y_pred))
```

5. Insights: The model accuracy is 84%

6. Hyperparameter tuning using GridSearchCV to see if the model performs better with exhaustive search over specified parameter values for an estimator.

```
# Import GridSearchCV
from sklearn.model_selection import GridSearchCV

# Define the grid of values for tol and max_iter
tol = [0.01,0.001, 0.0001]
max_iter = [100, 150, 200]

# Create a dictionary where tol and max_iter are keys and the lists of their values are corresponding values
param_grid = dict(tol = tol,max_iter = max_iter)
```

6.1 Finding the best performing model - by storing the best-achieved score and the respective best parameters

```
# Instantiate GridSearchCV with the required parameters
grid_model = GridSearchCV(estimator=logreg, param_grid= param_grid, cv=5)

# Use scaler to rescale X and assign it to rescaledX
rescaledX = scaler.fit_transform(X)

# Fit data to grid_model
grid_model_result = grid_model.fit(rescaledX, y)

# Summarize results
best_score, best_params = grid_model_result.best_score_, grid_model_result.best_params_
print("Best: %f using %s" % (grid_model_result.best_score_, grid_model_result.best_params_))
```
